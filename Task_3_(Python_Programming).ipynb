**Question 1**

Find Angle MBC

q4.PNG

```python
import math

AB= int(input())
BC = int(input())
CA = math.sqrt(AB**2 + BC**2)

CM = CA/2.0
adj = BC/2.0

output = int(round(math.degrees(math.acos(adj/CM))))

degree=chr(176)                                #for DEGREE symbol
print(str(output)+str(degree))
```
